```python
def format_string(x):
    if len(x) == 2:
        return ''
    elif x[0] == '[':
        return x[3:-2]
    else:
        return x
```

```python
recipe = recipe.applymap(format_string)
style = style.applymap(format_string)
fermentables = fermentables.applymap(format_string)
hops = hops.applymap(format_string)
yeasts = yeasts.applymap(format_string)
#miscs = miscs.applymap(lambda x: '' if len(x) == 2 else x[3:-2])
```

```python
def format_float(x):
    try:
        return np.float16(x)
    except:
        return x
```

```python
recipe = recipe.applymap(format_float)
style = style.applymap(format_float)
fermentables = fermentables.applymap(format_float)
hops = hops.applymap(format_float)
yeasts = yeasts.applymap(format_float)
#miscs = miscs.applymap(lambda x: '' if len(x) == 2 else x[3:-2])
```
